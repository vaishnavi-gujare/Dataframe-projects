```python
numeric_cols = [
    'Launch_Price_USD',
    'Launch_Price_INR',
    'Current_Price_USD',
    'Current_Price_INR',
    'Discount_Pct',
    'Rating',
    'Reviews_Count'
]
plt.figure(figsize=(18,12))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[col], bins=30, kde=True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
df_encoded = pd.get_dummies(df, drop_first=True)
print(df_encoded.head())
```

One-Hot Encoding was used to convert categorical values into numeric form.
Each category was converted into a separate column containing True/False (or 1/0) values.
This encoding helps machine learning models process categorical data correctly.
drop_first=True was used to avoid creating unnecessary duplicate information.
