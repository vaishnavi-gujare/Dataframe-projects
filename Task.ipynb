```python
plt.figure(figsize=(8,5))
sns.countplot(x='region', data=df)
plt.title('Number of Battles by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(df['year'], bins=10, kde=True)
plt.title('Distribution of Battle Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='attacker_size', y='defender_size', data=df)
plt.title('Attacker Size vs Defender Size')
plt.xlabel('Attacker Size')
plt.ylabel('Defender Size')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y='year', data=df)
plt.title('Box Plot of Year')
plt.ylabel('Year')
plt.show()
```

```python
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(
    df['battle_type'].value_counts(),
    labels=df['battle_type'].value_counts().index,
)
plt.title('Battle Type Distribution')
plt.show()
```
