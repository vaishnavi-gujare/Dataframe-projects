```python
numeric_cols = ['price', 'mileage', 'year', 'engineSize', 'mpg']
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 2, i)
    plt.hist(df[col], bins=20, edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['fuelType', 'transmission', 'model']
plt.figure(figsize=(18, 12))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3, 1, i)
    sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Count Plot of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

```python
# 1. Fuel Type: Petrol and Diesel vehicles are the most common.
# 2. Transmission: Manual transmission cars are more common than automatic cars.
# 3. Model: Some Ford models appear much more frequently than others.
# Market Trends: Customers prefer fuel-efficient and affordable vehicles.
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
print("Correlation with Price:")
print(corr_matrix['price'].sort_values(ascending=False))
```

```python
# Price shows a strong positive correlation with year. This means newer cars generally have higher prices.
# MPG may show a weak or moderate correlation with price.
# Cars with larger engines tend to be more expensive.
```

```python
# Dependent Feature (Target Variable):
target_feature = 'price'

# Independent Features (Input Variables):
input_features = [
    'model',
    'year',
    'transmission',
    'mileage',
    'fuelType',
    'tax',
    'mpg',
    'engineSize'
]

print("Dependent Feature (Target):", target_feature)
print("Independent Features (Inputs):", input_features)
# 1. Price is chosen as the dependent (target) feature because it is the value we want to predict or analyze.
# 2. The remaining columns act as independent features because they influence the selling price of a car.
# 3. Year affects price since newer cars generally cost more.
# 4. Mileage affects price because heavily used cars are usually cheaper.
```
